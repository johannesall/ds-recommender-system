1. How many unique users and movies are in the dataset?

```python
# 1. How many unique users and movies are in the dataset?
unique_users = ratings_df['userId'].nunique()
unique_movies = ratings_df['movieId'].nunique()

print(f'Unique users: {unique_users}')
print(f'Unique movies: {unique_movies}')
```

2. What is the average rating?

```python
# 2. What is the average rating?
average_rating = ratings_df['rating'].mean()
print(f'Average rating: {average_rating}')
```

3. Which rating appears the most?

```python
# 3. Which rating appears the most?
most_common_rating = ratings_df['rating'].mode().values[0]
print(f'Most common rating: {most_common_rating}')
```

4. Which movies were rated the best? Which movies were rated the worst?

```python
# 4. Which movies were rated the best? Which movies were rated the worst?
average_ratings = ratings_df.groupby('movieId')['rating'].mean()

# Find the movies with the highest average rating
boolean_max = (average_ratings == average_ratings.max())
best_movie_id = average_ratings[boolean_max].index.to_list()
print(f"Best movie ID: {best_movie_id}")
print(f"The number of best movies: {len(best_movie_id)}")
```

```python
# Find the movies with the lowest average rating
boolean_min = (average_ratings == average_ratings.min())
worst_movie_id = average_ratings[boolean_min].index.to_list()   
print(f"Worst movie ID: {worst_movie_id}")  
print(f"The number of worst movies: {len(worst_movie_id)}")
```

4. â­**Bonus**: Filter out movies that have been rated by less than 30 users. Do your results change?

```python
# â­**Bonus**: Filter out movies that have been rated by less than 30 users. Do your results change? 
ratings_count = ratings_df.groupby('movieId')['rating'].count()
ratings_count.head(10)
```

```python
# Filter out movies that have been rated by less than 30 users
boolean_30 = (ratings_count >= 30)
average_ratings_30 = average_ratings[boolean_30]
average_ratings_30
```

```python
# Which movies were rated the best? Which movies were rated the worst?
best_movie_id_30 = average_ratings_30[average_ratings_30 == average_ratings_30.max()].index.to_list()
worst_movie_id_30 = average_ratings_30[average_ratings_30 == average_ratings_30.min()].index.to_list()
# Print the results
print(f"Best movies ID with at least 30 ratings: {best_movie_id_30}")
print(f"The number of best movies with at least 30 ratings: {len(best_movie_id_30)}")
print(f"Worst movies ID with at least 30 ratings: {worst_movie_id_30}")
print(f"The number of worst movies with at least 30 ratings: {len(worst_movie_id_30)}")
```

5. How are the ratings distributed?

```python
# 5. How are the ratings distributed?
relative_frequency = ratings_df['rating'].value_counts(normalize=True).sort_index()
relative_frequency
```

5. â­**Bonus**:: Plot the relative frequency table using a bar plot.

```python
relative_frequency.plot(kind='bar', color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Relative frequency')
plt.title('Relative frequency of ratings')
plt.xticks(rotation=0)
plt.show()
```

6. Count the number of ratings per user.

```python
# 6. Count the number of ratings per user.
ratings_per_user = ratings_df.groupby(by='userId')['rating'].count().sort_values(ascending=False)
ratings_per_user
```

```python
# How many users have rated more than 20 movies?
count_user_above_20 = (ratings_per_user > 20).sum()
print(f'Number of users who have rated more than 20 movies: {count_user_above_20}')
```

```python
# How many movies rated the user with least number and most number of ratings?
min_ratings = ratings_per_user.min()
max_ratings = ratings_per_user.max()
print(f'Minimum number of ratings: {min_ratings}')
print(f'Maximum number of ratings: {max_ratings}')
```

```python
# Plot the distribution of the number of ratings per user.
plt.figure(figsize=(10, 5))
sns.histplot(ratings_per_user, bins=30, color='skyblue')
plt.xlabel('Number of ratings')
plt.yscale('log')
#plt.xscale('log')
plt.ylabel('Frequency')
plt.title('Distribution of the number of ratings per user')
plt.show()
```

```python
# boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(ratings_per_user, color='skyblue')
plt.xlabel('Number of ratings')
plt.title('Distribution of the number of ratings per user')
plt.show()
```

How would you caracterize the distribution?

The distribution of the number of ratings per user is right-skewed. Most users have rated less than 100 movies, and a few users have rated more than 100 movies.
